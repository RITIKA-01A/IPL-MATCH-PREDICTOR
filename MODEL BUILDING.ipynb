```python
total_score_df = delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score_df
```

```python
#adding the total runs in the match dataframe
match_df = match.merge(total_score_df[['match_id','total_runs']] ,left_on = 'id' , right_on = 'match_id')
match_df.head(2)
```

```python
match_df['team1'].unique()
```

```python
#list of teams currently playing ipl
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'   
]
```

```python
#replacing the name of the team with the recent  names of the team
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')


match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#ensuring that the team consists only those teams that are currently playing ipl
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
print(match_df['team1'].nunique())
print(match_df['team2'].nunique())
```

```python
match_df.shape
```

THE MATCHES HAVING THE INVALID TEAMS ARE ELIMINATED

```python
match_df.head()
```

```python
match_df['dl_applied'].value_counts()
```

15 matches are rain affected

```python
#we want those matches where there is no rain in the data

match_df = match_df[match_df['dl_applied']==0]
match_df.shape
```

```python
match_df
```

```python
#features we need in the match_df data


#batting_team,bowling_team , city , runs_left , balls_left , wickets_left , total_runs_x , crr , rrr , result
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery , on = 'match_id' )
```

```python
#we only want the second inning data
delivery_df = delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.head()
```

```python
delivery_df.shape
```

```python
type(delivery_df['total_runs_y'])
```

```python
# Using transform instead of cumsum directly
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].transform(pd.Series.cumsum)

```

```python
delivery_df
```

```python
#finding the runs left to score in each ball
#total_runs_x  -->> total runs to score(overall)
#current_score -->> runs scored till now
##(total_runs_x - current_score = runs left to score )

delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
#finding how may balls are left after each ball
#there are total 21 overs in the match
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
delivery_df['player_dismissed'].value_counts()
```

```python
#wickets left
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x =="0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('match_id')['player_dismissed'].transform(pd.Series.cumsum)
#wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 -wickets
delivery_df
```

```python
#finding current run rate

delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df
```

```python
#finding required run rate'
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
#extracting the columns needed
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df 
```

```python
#shuffling the data to avoid any kind of bias
final_df = final_df.sample(final_df.shape[0])
```

```python
#checking for null value
final_df.isnull().sum()
```

```python
#dropping the rows with missing value
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] !=0]
```

```python
final_df.sample()
```

# MODEL BUILDING

```python
X = final_df.drop(columns=['result'])
y = final_df['result']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# CATEGORICAL TO NUMERICAL

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder = 'passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1' , trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```

```python
pipe.predict_proba(X_test)[7]
```
